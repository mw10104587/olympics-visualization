```python
# data = pd.read_csv('generated data/formated-all-1896-2012.csv')
data = pd.read_csv('data/summer-olympics-1896-2012.csv', encoding="ISO-8859-1")
```

<h1> This paart only focus on female weightlifting</h1>

```python
wl_female = data[(data.Sport == 'Weightlifting') & (data['Event gender'] == 'W')]
wl_female.shape
wl_female
```

```python
female_wl_gold = wl_female[wl_female.Medal == 'Gold']
female_wl_gold = female_wl_gold.groupby(['NOC']).count()[['Medal']]
female_wl_gold['pct'] = np.around( list(female_wl_gold.Medal/np.sum(female_wl_gold.Medal)), decimals=4) 
female_wl_gold['pct'] = female_wl_gold['pct'] * 100
female_wl_gold.sort_values(['pct'], ascending=False)
```
